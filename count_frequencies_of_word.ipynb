```python
#FIRSTLY WE WILL DEFINE A FUNCTION AND GIVE FILE PATH IN PARAMETERS.
def counting_frequency(file_path):
    #Now we will make a empty dictionary to store word and its frequency in it.
    word_and_count={}
    # Now here we will open the file and read it from given file path
    with open(file_path,'r') as file:
        #Now we will apply for loop in lines to itrate in line in file
        for lines in file:
            #we will split lines for words
            words=lines.split()
# Loop through each word, convert it to lowercase, remove punctuation,
# and count its frequency using a dictionary
            for word in words:
                word=word.lower().strip(".,!?;:'\"")
                word_and_count[word]=word_and_count.get(word,0)+1

    return word_and_count
```

```python
filepath='frequency.txt'
answer=counting_frequency(filepath)
print(answer)
```
